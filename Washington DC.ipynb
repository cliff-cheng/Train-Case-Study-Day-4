```python
import pandas as pd
import numpy as np 
import os
import folium as fl
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline 
```

```python
city = pd.read_csv('data/cities.csv')
city = city.drop(['url_name', 'country_state'], axis = 1)

track = pd.read_csv('data/tracks.csv')
track = track.drop(['closure'], axis = 1)

station = pd.read_csv('data/stations.csv')
station = station.drop(['closure'], axis = 1)

lines = pd.read_csv('data/lines.csv')
lines = lines.drop(['transport_mode_id'], axis = 1)
track
```

```python
city_wash = city.loc[city['name'] == 'Washington']
city_coords = city_wash.iloc[0]['coords']
city_coords = city_coords.replace('POINT', '')
city_coords = city_coords.replace('(', '')
city_coords = city_coords.replace(')', '')
city_coords = city_coords.split(' ')    
```

```python
city_cor_list = []
for x in city_coords:
    city_cor_list.append(float(x))
city_cor_list.reverse()
```

```python
track_wash = track.loc[track['city_id'] == 252]
track_wash_coor = track_wash.iloc[:]['geometry']
track_coor_list = []

for x in track_wash_coor:
    y = x.replace('LINESTRING', '').replace('(', '').replace(')', '')
    z = y.split(',')
    track_coor_list.append(z)
track_coor_list
```

```python
track_wash = track.loc[track['city_id'] == 252]
track_wash_coor = track_wash.iloc[:]['geometry']
track_coor_list = []

for x in track_wash_coor:
    y = x.replace('LINESTRING', '').replace('(', '').replace(')', '')
    z = y.split(',')
    ab = []
    for a in z:
        a = a.split(' ')
        a = (float(a[1]), float(a[0]))
        ab.append(a)
    track_coor_list.append(ab)
    
track_coor_list
```

```python
map_osm = fl.Map(location=city_cor_list, tiles = 'Stamen Toner', zoomstart = 12)

for x in track_coor_list:
    for y in x:
        fl.Marker(y).add_to(map_osm)
map_osm

```
